```python
# If not change the data types of some columns to what they are supposed to be

car_df.Latest_Launch = pd.to_datetime(car_df.Latest_Launch)

```

```python
# Are there any missing values in the dataset? If so, how many and in which columns?


car_df.isnull().sum()
```

```python
# Fill the missing values in the year_resale_value with the average year_resale_value, then drop all other rows with missing values in the dataset

car_df["__year_resale_value"].fillna(car_df["__year_resale_value"].mean(), inplace=True)
car_df.dropna(inplace=True)

```

```python
# Clean the name of the __year_resale_value column by removing the leading underscore
car_df.rename(columns={"__year_resale_value": "year_resale_value"}, inplace=True)


```

## **Specific Analysis:**

```python
# Find out the most common car brand in the dataset using value_counts()
car_df["Manufacturer"].value_counts()
```

```python
# Which manufacturer has the highest and lowest average sales volume?

car_df.groupby("Manufacturer")["Sales_in_thousands"].mean().sort_values(ascending=False)
```

```python
# What is the distribution of car prices in the dataset?
import matplotlib.pyplot as plt

car_df["Price_in_thousands"].plot(kind="hist")
plt.xlabel("Price in thousands")
plt.ylabel("Frequency")
plt.title("Distribution of car prices");
```

```python
# Plot a correlation matrix to see if different numerical variables in the dataset correlate with each other
import seaborn as sns



corr_matrix = car_df[["Sales_in_thousands",	"year_resale_value",	"Price_in_thousands",	"Engine_size",	"Horsepower",	"Wheelbase",	"Width",	"Length",	"Curb_weight",	"Fuel_capacity",	"Fuel_efficiency",	"Latest_Launch",	"Power_perf_factor"]].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(data=corr_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show();
```

```python
# Plot a histogram of the car prices using Matplotlib.

car_df.Price_in_thousands.plot(kind="hist")
plt.xlabel("Price in thousands")
plt.ylabel("Frequency")
plt.title("Distribution of car prices");
```

```python
# Plot the relationship between price and latest launch year.

sns.scatterplot(data=car_df, x="Latest_Launch", y="Price_in_thousands")
plt.xlabel("Latest Launch Year")
plt.ylabel("Price in Thousands")
plt.title("Relationship between Price and Latest Launch Year");
```

```python
# Group cars by manufacturers and analyze their average price and fuel efficiency.

car_df.groupby("Manufacturer")[["Price_in_thousands", "Fuel_efficiency"]].mean()
```

```python
# Can you identify and visualize the most popular car brands?

car_df.Manufacturer.value_counts().plot(kind="bar")
plt.xlabel("Manufacturer")
plt.ylabel("Count")
plt.title("Most Popular Car Brands");
```

```python
# Which car model has the highest resale value compared to its initial price?

car_df["Resale_Value_Percentage"] = (car_df["year_resale_value"] / car_df["Price_in_thousands"]) * 100
car_df.sort_values("Resale_Value_Percentage", ascending=False).head(1)
```

```python
# Find the top 3 fuel-efficient cars with an engine size above 2.5 liters.

car_df[car_df["Engine_size"] > 2.5].sort_values("Fuel_efficiency", ascending=False).head(3)
```

```python
# Which Lexus model has the highest horsepower?

car_df[car_df["Manufacturer"] == "Lexus"].sort_values("Horsepower", ascending=False).head(1)
```
